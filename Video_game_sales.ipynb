##### VGChartz_Score seems to be repetitive and needs to be removed along with a few other variables that do not seem to be useful. 

```python
data = data_all.drop(['Rank','Last_Update','url','status','img_url','VGChartz_Score'], axis=1)
```

```python
data.describe(include='all')
```

### Top sellers based on publisher company

```python
total_sale_company=data.groupby('Publisher').sum()
```

```python
top_companies = total_sale_company.sort_values (by=['Global_Sales'], ascending=False)
```

```python
top_companies[:15].Global_Sales.plot(kind='bar', title="Top 15 video game publishers based on global sales")
```

### Top sellers based on genre

```python
total_sale_genre=data.groupby('Genre').sum()
```

```python
top_genre = total_sale_genre.sort_values (by=['Global_Sales'], ascending=False)
```

```python
top_genre[:15].Global_Sales.plot(kind='bar', title="Top 15 video game genre based on global sales")
```

### Top sellers based on platform

```python
total_sale_platform=data.groupby('Platform').sum()
```

```python
top_platform = total_sale_platform.sort_values (by=['Global_Sales'], ascending=False)
```

```python
top_platform[:15].Global_Sales.plot(kind='bar', title="Top 15 favorite platforms according to global sales")
```

### Investigating which score correlates with total sell

```python
import seaborn as sns
```

```python
sns.pairplot(data[['User_Score','Critic_Score','ESRB_Rating','Vgchartzscore','Global_Sales','NA_Sales','PAL_Sales','JP_Sales']])
```

#### Although based on this data none of the scores seems to be linearly correlated with sales, however, User_Score, Critics_Score and Vgchartzscore seem to be correlated. This means that in general professional critics agree with users experience

### Investigating the most favorite game genre based on critics score

```python
avg_Critics_score = data.groupby('Genre').mean().Critic_Score.sort_values(ascending=False)
```

```python
avg_Critics_score[:10].plot(kind='bar', title="Top 10 game genre based on critics scores", colormap='gray')
```

```python
avg_User_score = data.groupby('Genre').mean().User_Score.sort_values(ascending=False)
```

```python
avg_User_score[:10].plot(kind='bar', title="Top 10 game genre based on user scores", colormap='gray')
```

#### Although user experience and professional critics are slightly different among different genre, there is no significant difference among the different genre in general.
